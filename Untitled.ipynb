```python
import os
import glob
import re
import pandas as pd
```

```python
# Reading csv files from ETF_data folder
csv_files = glob.glob(os.path.join("ETF_data", "*.csv"))
df_list = []

# For each csv file extract the ETF name using regex and add the name to the a new column called "ETF"
# Add each csv file to df_list
for file in csv_files:
    df = pd.read_csv(file)
    match = re.search(r"/([^/]+)\.csv$", file)
    if match:
        ETF_str = match.group(1)
        df["ETF"] = ETF_str
    df_list.append(df)

# Remove any empty data frames
df_list = [df for df in df_list if not df.empty]

# Combine all data frames into one master data frame
master_df = pd.concat(df_list, ignore_index=True)
master_df
```

```python
# Data inspection
print(master_df.shape)
print(master_df.columns.tolist())
master_df.head()
```

```python
summary_stats = master_df.groupby('ETF').describe().round(2)
print("\nSummary Statistics by ETF:")
summary_stats
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
etf_names = master_df['ETF'].unique()
for etf in etf_names:
    if (etf == "VOO"):
        df_etf = master_df[master_df['ETF'] == etf].copy()
        df_etf.sort_values('Date', inplace=True)
        
        plt.figure(figsize=(8, 4))
        plt.plot(df_etf['Date'], df_etf['Adj Close'])
        plt.title(f"Closing Price Over Time - {etf}")
        plt.xlabel("Date")
        plt.ylabel("$/Unit")
        ax = plt.gca()
        ax.xaxis.set_major_locator(MaxNLocator(7))
        plt.tight_layout()
        plt.show()
```

```python
master_df.sort_values(by=['ETF','Date'], inplace=True)

# Group by ETF, then compute daily returns
master_df['Daily_Return'] = master_df.groupby('ETF')['Close'].pct_change()

# Quick check
print("\nDaily Returns preview:")
master_df[['ETF','Date','Close','Daily_Return']].head(10)
```

```python
for etf in etf_names:
    if (etf == "QQQM"):
        df_etf = master_df[master_df['ETF'] == etf].copy()
        
        plt.figure(figsize=(8,4))
        df_etf['Daily_Return'].hist(bins=50)
        plt.title(f"Distribution of Daily Returns - {etf}")
        plt.xlabel("Daily Return")
        plt.ylabel("Frequency")
        plt.tight_layout()
        plt.show()
```
