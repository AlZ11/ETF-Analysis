```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import itertools
import statsmodels.api as sm
import plotly.graph_objects as go
from matplotlib.ticker import MaxNLocator
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima.model import ARIMA
```

# Load data

```python
master_df = pd.read_pickle("master_df.pkl")
master_df
```

# ARIMA orders

```python
# Change this to see analysis for a particular ETF
etfKey = "SPY"
etfDF = (master_df.loc[master_df["ETF"] == etfKey, ["Date", "Adj Close"]].sort_values("Date").reset_index(drop=True))
```

```python
# Lets check for stationarity
plt.figure(figsize=(12, 6))
plt.plot(etfDF["Date"], etfDF["Adj Close"])
plt.xlabel("Date")
plt.ylabel("Adj Close")
plt.title(f"{etfKey} Date VS Adj Close")
plt.show()
```

```python
# Compute daily log returns:  r_t = ln(P_t / P_{t-1}) to transform to stationary
etfDF["logRet"] = np.log(etfDF["Adj Close"]).diff()

# Drop the first NaN created by .diff()
etfDF = etfDF.dropna(subset=["logRet"])
etfDF.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(etfDF["Date"], etfDF["logRet"])
plt.xlabel("Date")
plt.ylabel("Log return")
plt.title(f"{etfKey} Date VS Log return")
plt.show()
```

```python
# Augmented Dickey Fuller test to verify stationarity
series = etfDF[["Date", "logRet"]].copy().set_index("Date").sort_index()["logRet"].dropna()

adf_stat, p_val, lags, nobs, crit, icbest = adfuller(series, autolag="AIC")

print(f"ADF statistic  : {adf_stat: .4f}")
print(f"p-value        : {p_val: .4f}")
print(f"lags used      : {lags}")
print(f"obs in regress.: {nobs}")
print("critical values:")
for k, v in crit.items():
    print(f"   {k}% => {v: .4f}")

# Confirms stationarity
alpha = 0.05
print("\nReject H0: series is likely STATIONARY") if p_val < alpha else print("\nFail to reject H0: series NON-stationary.")
```

```python
y = etfDF["logRet"]
# Silence convergence warnings from statsmodels for cleaner output
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)

bestAIC = np.inf
bestP, bestQ = 0, 0

# 36 combinations for p & q
for p in range(6):
    for q in range(6):
        try:
            res = sm.tsa.statespace.SARIMAX(y, order=(p, 0, q), trend="n").fit(disp=False)
            if res.aic < bestAIC:
                print(res.aic, p, q)
                bestAIC, bestP, bestQ = res.aic, p, q
        except (ValueError, np.linalg.LinAlgError):
            # skip non-invertible combos
            continue

print(f"Best ARMA order by AIC: p={bestP}, q={bestQ}  (d=0)")
```

```python
# The model suggests that the order of the AR model is 0, lets confirm this through a PACF graph to see whether any lags are actually statistically significant
series = etfDF["logRet"].dropna()

# PACF plot
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot()

plot_pacf(series, lags=30, method="ywm", ax=ax)  # Yule-Walker modified
ax.set_title(f"Partial Autocorrelation Function (PACF) â€“ {etfKey} Daily Log Returns")
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Fit ARIMA(p,q) on the entire log return series
y = etfDF[["Date","logRet"]].dropna()
warnings.filterwarnings("ignore", category=UserWarning)
armaFull = sm.tsa.statespace.SARIMAX(y["logRet"], order=(bestP, 0, bestQ), trend="n").fit(disp=False)

print(armaFull.summary())      # full stats table
```

```python
h = 200           # predict the next 20 trading days
# ------------------------------------------------------------------
# 1.  In-sample fitted log-returns  (modelâ€™s best guess for data it saw)
# ------------------------------------------------------------------
fitted_log = armaFull.fittedvalues            # aligns with training index

# ------------------------------------------------------------------
# 2.  Out-of-sample forecast log-returns  (h steps ahead)
# ------------------------------------------------------------------
fc_res   = armaFull.get_forecast(steps=h)
fc_log   = fc_res.predicted_mean             # Series length = h
ci_log   = fc_res.conf_int(alpha=0.05)       # lower/upper log-return bounds

# ------------------------------------------------------------------
# 3.  Undo the log-return transform to get *price* paths
#     Helper: cumulative-sum + exp, then scale by a starting price
# ------------------------------------------------------------------
def logret_to_price(start_price: float, log_ret: pd.Series) -> pd.Series:
    """
    Convert a series of log-returns into a price series that starts at start_price.
    """
    return start_price * np.exp(log_ret.cumsum())

# --- Figure out the anchor price safely -----------------------------------
first_idx = y.index[0]                           # index label of first training row
anchor_price = etfDF["Adj Close"].shift(1).loc[first_idx]

# --- Reconstruct in-sample fitted prices ---------------------------------
cum_log_ret = armaFull.fittedvalues.cumsum()     # ln P_t âˆ’ ln P_anchor
fitted_price = anchor_price * np.exp(cum_log_ret)

# 3-b  Forecast price path
last_obs_price   = etfDF["Adj Close"].loc[y.index[-1]]       # price just before forecast
price_forecast   = logret_to_price(last_obs_price, fc_log)

# 3-c  95 % price band (same transformation on lower/upper log-return bounds)
price_lo = logret_to_price(last_obs_price, ci_log.iloc[:, 0])
price_hi = logret_to_price(last_obs_price, ci_log.iloc[:, 1])

# ------------------------------------------------------------------
# 4.  Quick look
# ------------------------------------------------------------------
print("Next-day log-return forecast:", fc_log.iloc[0])
print("Next-day price forecast:    ", price_forecast.iloc[0])
```

```python
# Split train and test set
train = y.iloc[:-int(len(y) * 0.2)]
test = y.iloc[-int(len(y) * 0.2):]

# Build ARIMA model and inverse the natural log
model = ARIMA(train["logRet"], order=(bestP, 0, bestQ)).fit()
logForecasts = model.forecast(len(test))
#logForecasts = armaFull.predict(len(test))
lastPrice = etfDF["Adj Close"].iloc[len(train)]
cumLogRet = logForecasts.cumsum()
priceForecast = lastPrice * np.exp(cumLogRet)
```

```python
def plot_forecasts(forecasts: list[float], title: str) -> None:
    """Function to plot the forecasts."""
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=y["Date"], y=etfDF["Adj Close"], name='Train'))
    fig.add_trace(go.Scatter(x=test["Date"], y=lastPrice * np.exp(test["logRet"].cumsum()), name='Test'))
    fig.add_trace(go.Scatter(x=test["Date"], y=priceForecast, name='Forecast'))
    fig.update_layout(template="simple_white", font=dict(size=18), title_text=title,
                      width=1300, title_x=0.5, height=800, xaxis_title='Date',
                      yaxis_title='Price')

    return fig.show()
plot_forecasts(priceForecast, f"ARIMA forecast prices on {etfKey}")
```

```python
import plotly.graph_objects as go
import numpy as np

def plot_forecasts(
        train_dates, train_price,
        fitted_price,                 # in-sample fit (same index as train)
        test_dates,  test_price,
        forecast_price,               # out-of-sample prediction (same index as test)
        title:str = ""
    ) -> None:
    """
    Plot actual vs fitted vs forecast prices.

    Parameters
    ----------
    train_dates, train_price : pandas Series/array-like
        Dates and actual adjusted closes used for training.
    fitted_price : pandas Series/array-like
        Model-reconstructed prices over the *training* window.
    test_dates, test_price : pandas Series/array-like
        Dates and actual prices in the held-out window.
    forecast_price : pandas Series/array-like
        Modelâ€™s point forecast converted back to price level.
    title : str
        Title for the figure.
    """
    fig = go.Figure()

    # 1. actual training price
    fig.add_trace(go.Scatter(
        x=train_dates, y=train_price,
        name="Train (actual)", line=dict(color="steelblue")
    ))

    # 2. fitted (in-sample) price
    fig.add_trace(go.Scatter(
        x=train_dates, y=fitted_price,
        name="Train (fitted)", line=dict(color="orange", dash="dot")
    ))

    # 3. actual test price
    fig.add_trace(go.Scatter(
        x=test_dates, y=test_price,
        name="Test (actual)", line=dict(color="green")
    ))

    # 4. forecast price
    fig.add_trace(go.Scatter(
        x=test_dates, y=forecast_price,
        name="Forecast", line=dict(color="red", dash="dash")
    ))

    fig.update_layout(
        template="simple_white",
        font=dict(size=16),
        title_text=title,
        title_x=0.5,
        width=1200,
        height=600,
        xaxis_title="Date",
        yaxis_title="Price"
    )
    fig.show()
```

```python
# --- pieces you already have from previous steps -----------------
train_dates   = etfDF.loc[train.index, "Date"]
train_price   = etfDF.loc[train.index, "Adj Close"]
fitted_price  = fitted_price                     # from earlier snippet
test_dates    = etfDF.loc[test.index, "Date"]
test_price    = lastPrice * np.exp(test["logRet"].cumsum())
forecast_price = price_forecast                  # Series built from log-return forecast

# --- draw the chart ---------------------------------------------------------
plot_forecasts(
    train_dates, train_price,
    fitted_price,
    test_dates,  test_price,
    forecast_price,
    title=f"ARIMA fitted & 20-day forecast on {etfKey}"
)
```
